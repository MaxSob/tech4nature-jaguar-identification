```python
def changelabel(label):
    if label == 'jaguar':
        return 1
    else: 
        return 0
```

```python
df['new_label'] = df['label'].apply(changelabel)
```

```python
from tensorflow.keras.preprocessing.image import img_to_array 
from tensorflow.keras.preprocessing.image import load_img
```

```python
def data(path):
    image = load_img(path, target_size=(128, 128))
    image = img_to_array(image)
    return image
```

```python
df['data'] = df['name'].apply(data)
```

```python
# convert from the range [0, 255] to [0, 1] 
dataX = np.array(list(df['data']), dtype="float32") / 255.0
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D, GlobalAveragePooling2D, Flatten,  BatchNormalization

model = Sequential()                                 # Linear stacking of layers

# Convolution Layer 1
model.add(Conv2D(32, (3, 3), input_shape=(128,128,3))) # 32 different 3x3 kernels -- so 32 feature maps
model.add(BatchNormalization(axis=-1))               # normalize each feature map before activation
convLayer01 = Activation('relu')                     # activation
model.add(convLayer01)

# Convolution Layer 2
model.add(Conv2D(32, (3, 3)))                        # 32 different 3x3 kernels -- so 32 feature maps
model.add(BatchNormalization(axis=-1))               # normalize each feature map before activation
model.add(Activation('relu'))                        # activation
convLayer02 = MaxPooling2D(pool_size=(2,2))          # Pool the max values over a 2x2 kernel
model.add(convLayer02)

# Convolution Layer 3
model.add(Conv2D(64,(3, 3)))                         # 64 different 3x3 kernels -- so 64 feature maps
model.add(BatchNormalization(axis=-1))               # normalize each feature map before activation
convLayer03 = Activation('relu')                     # activation
model.add(convLayer03)

# Convolution Layer 4
model.add(Conv2D(64, (3, 3)))                        # 64 different 3x3 kernels -- so 64 feature maps
model.add(BatchNormalization(axis=-1))               # normalize each feature map before activation
model.add(Activation('relu'))                        # activation
convLayer04 = MaxPooling2D(pool_size=(2,2))          # Pool the max values over a 2x2 kernel
model.add(convLayer04)
model.add(Flatten())                                 # Flatten final 4x4x64 output matrix into a 1024-length vector

# Fully Connected Layer 5
model.add(Dense(128))                                # 512 FCN nodes
model.add(BatchNormalization())                      # normalization
model.add(Activation('relu'))                        # activation

# Fully Connected Layer 6                       
model.add(Dropout(0.5))                              # 20% dropout of randomly selected nodes
model.add(Dense(1))                                 # final 10 FCN nodes
model.add(Activation('sigmoid'))    
```

```python
# convert from the range [0, 255] to [0, 1] 
dataX = np.array(list(df['data']), dtype="float32") / 255.0

labels = list(df['new_label'])

split = train_test_split(dataX, labels, test_size=0.20, random_state=42)


(trainImages, testImages) = split[:2]
(trainLabels, testLabels) = split[2:4]

INIT_LR = 1e-4
opt = Adam(INIT_LR)

model.compile(loss='binary_crossentropy',
              optimizer=opt,
              metrics=['accuracy','Recall', 'Precision', 'FalsePositives', 'TruePositives', 'FalseNegatives'])
```

```python
train_x = np.asarray(trainImages)
train_y = np.asarray(trainLabels)
validation_x = np.asarray(testImages)
validation_y = np.asarray(testLabels)
```

```python
history = model.fit(train_x, train_y, validation_data=(validation_x, validation_y), batch_size=16, epochs=100, verbose=1)
                   #steps_per_epoch = 100)
                   #callbacks = [DisplayCallback()])
```

```python
loss, accuracy, recall, precision, fp, tp, fn = model.evaluate(train_x,train_y)
print('Test accuracy :', accuracy)
print('Test recall :', recall)
print('Test precision :', precision)
print('False Positive :', fp)
print('True Positive :', tp)
print('False Negative :', fn)
tn = len(train_x)-fp-tp-fn
print('True Negative :', tn)
print('F1 :', (2*(precision*recall))/(precision + recall))
```
